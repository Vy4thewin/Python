### Wrangling time| Let's search for data positions

For our topic, we only want positions related to data analyst or data sceintist. Let's see if these positions are avaialble with our current database with a like search. 

First, I want our rough draft data frame to only have columns relevant to our search, we can rename all the unnessasary columna as the current column names are clear and direct. We can resize all the unwanted column names to a shorter nickname. 


Moving on, all the column types from the json connection might have been remove in the retreival. We can reconvert column values depending on its type. For example, date was not recongized in the transmission over so we can reassign with .to_datetime().

```python
# Modify multiple column names.
rd=data.rename(columns={"residency_requirement":"Res_requir","civil_service_title":"cs_title","division_work_unit":"dv_work"})
# Show the rename work
print(rd["Res_requir"].head())

# Look at the structure of your data – are any variables improperly coded? Such as strings or characters? Convert to correct structure if needed.
#USing to_datetime to reconvert the values back to date incase we want to filter for more recent positions
rd["posting_date"]=pd.to_datetime(rd["posting_date"])
rd["posting_updated"]=pd.to_datetime(rd["posting_updated"])
rd["process_date"]=pd.to_datetime(rd["process_date"])
rd["post_until"]=pd.to_datetime(rd["post_until"])

#check if the posting are now in the correct format
rd["posting_date"].info()
```

For some data clean up, there are certain positions where the shift hours are unavailable due it it not being listed or the process of the data collection did not retrieve the information. We can fill the missing shift hours with "Not Available" with .fillna option.

We can create two new columns of a calculated rate of pay from our min/max annual salaries. In order to calculate the rate of pay, we can use the formula Hourly = Salary per year / (Hours per week × Weeks per year) [^2]. We can assume that each poistion alloates 40 hours per week to avoid calculating over pay and that the minimum salary accounts after tax. We can create the column "Hour_rate" in our data frame.

[^2]:https://www.adp.com/resources/tools/calculators/hourly-paycheck-calculator.aspx

```python
# Fix missing and invalid values in data.
rd["hours_shift"].fillna("Not Available",inplace=True)

# Check whether numeric values are present in a given column of your dataframe.
#Lets check that there's only float values in the salary columns
print(rd["salary_range_from"].dtype)
print(rd["salary_range_to"].dtype)

# Create new columns based on existing columns or calculations.
#caculating the rate of pay assuming a person works a 40 hr work week on average
rd["Hour_rate_min"]=rd["salary_range_from"]/(40*52)
rd["Hour_rate_max"]=rd["salary_range_to"]/(40*52)

#Checking out our new columns
print(rd[["cs_title","salary_range_from","Hour_rate_min","Hour_rate_max"]].head())

```

#### Reducing and refining our scope| Only data specific jobs

Heading down our job board, we can move to dropping some rows that are not of use to us. To reduce the size of our table, we can drops the columns "posting_type,number_of_positions,title_classification,title_code_no". We can also drop any positions that do not display their salaries or if its not annually as we want to weed out additional time to learn if the position's salary is enough to pursue the long application/ interview process. Then, we can now focus our search to our interest.

The amount of ways to search for Data analyst, data scientist, and data engineer is tricky as using the full term may be too restrictive. We can se using terms like data, analyst, scientist, engineer makes the search broad enough to get positions related to our interests. In our jobs table, this data frame contains all positions that's of interest.

```python
# Drop column(s) from your dataset.
rd.drop(["posting_type","number_of_positions","title_classification","title_code_no"],axis=1)

# Drop a row(s) from your dataset.
#drop rows that do not listed a salary amounts or it is not a annual salary as its short term/ low wage
rd=rd[rd["salary_range_from"].notnull()]
rd=rd[rd["salary_frequency"]=="Annual"]
rd["business_title"].count()

# Filter your data based on some condition.

# Convert all the string values to upper or lower cases in one column.
#Before we can filter, bc str contains is case sensitive, lets transform of job titles to lower case
rd["business_title"]=rd["business_title"].str.lower()
rd["cs_title"]=rd["cs_title"].str.lower()

#We can use str contains to search for the terms below without it being a exact match with the regex match, it will open the match to searhc for positions like Senior Data analyst|data analyst w/o writing for all terms
#adding data engineer as the only full term as engineer in this 
jobs=rd[rd["business_title"].str.contains("data|analyst|scientist|engineer|data engineer",regex=True)]
jobs.head()

```

#### What's in our market?

Now, we can get into the focus of this data frame. Let's sort our potential jobs by career level and hourly rate to get into the most high seeking positions. We can group these positions by their career level and find its summary statistics with a .agg and retreieve the Hourly rate avg,min, and max.

```python
# Sort your data based on multiple variables.
#Sort popening by the career level and its hourly rate
jobs.sort_values(["career_level","hours_shift"])

# Group your dataset by one column, and get the mean, min, and max values by group.
#Group all position by its career level and find the mean,min, and max
jobs.groupby("career_level").agg({'salary_range_from':['mean','min','max']})
```

```python
# Group your dataset by two columns and then sort the aggregated results within the groups.
# Let's see if we can also find the the average salary based on position with our data
jobs.groupby(["career_level","cs_title"]).agg({'Hour_rate_min':['mean','min','max']})
```

```python
#Look into summary stats for openings for entry level positiosn with je
je=jobs[jobs["career_level"]=="Entry-Level"]
je.groupby(["career_level","business_title","cs_title"]).agg({'Hour_rate_min':['mean','min','max']})
```

```python
#checking to make sure that data type jobs exisits 
jd=jobs[jobs["business_title"]=="data analyst"]
js=jobs[jobs["business_title"]=="data scientist"]
jd.groupby(["career_level","business_title","cs_title"]).agg({'Hour_rate_min':['mean','min','max','count']})
js.groupby(["career_level","business_title","cs_title"]).agg({'Hour_rate_min':['mean','min','max','count']})
```

We can create a list of most frequent terms to see if its more than # of working expeirence or if a master's degree can substitue in. I noticed that the qualifications are seperated by commas in the string, so we can break out the string into into its own column.

```python
#Lets be more specific in our clause statement to see most seek out qualifications/skills
import re
import string
import nltk
from  nltk.corpus import  names, words, stopwords
import numpy as np

#refinig the scope for positions that only contain data analyst or data scientist in the title
jobs_re=rd[rd["business_title"].str.contains("data analyst| data scientist|data engineer",regex=True)]

jobs_words=jobs_re["minimum_qual_requirements"].str.split(",")
words=jobs_words.tolist()

#Convert from a nest list to a flat list to count word frequency
words=list(np.concatenate(words))

#Count the # of times a phrase has come up in the minimum requirements and print out the top 20 results
freq_skill=sorted(nltk.FreqDist(words).items(),key=lambda x: x[1], reverse=True)
print(freq_skill[0:10])

#Noticed alot of phrase has unnesssary words, lets break the phrases down further and remove all stop words
splt_words=[i.split() for i in words]
splt_words=list(np.concatenate(splt_words))
```

```python
#import in the stop words list and use a lower() as its case sensitive for the for loop
#only included words that are not apart of the stop list 
r=stopwords.words('english')
cw=[n for n in splt_words if n.lower() not in r]

skills=sorted(nltk.FreqDist(cw).items(),key=lambda x: x[1], reverse=True)
print(skills[0:20])
```
